```python
import keras
from keras.models import Sequential, Model
from keras.layers import Input
from keras.preprocessing.sequence import pad_sequences
from keras import optimizers
from src.prepossessing import prepossessing
from src.MemN2N import MemN2N


path_1k = '/home/ro/dataset/tasks_1-20_v1-2/en/'
path_10k = '/home/ro/dataset/tasks_1-20_v1-2/en-10k/'
```

```python
%%time
for task in range(20) :
    trainS, trainQ, trainA, testS, testQ, testA, max_story_size, sentence_size, vocab_size = prepossessing(task + 1, path_1k)
    embedding_size = 20
    story = Input(shape=(max_story_size, sentence_size,))
    question = Input(shape=(sentence_size,), dtype='int32')
    answer = MemN2N(max_story_size, sentence_size, vocab_size, embedding_size, 3)([story, question])
    model = Model([story, question], answer)
    model.compile(optimizer='RMSprop', loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    model.fit([trainS, trainQ], trainA, batch_size=32, epochs=100, verbose=0, validation_data=([testS, testQ], testA))
    print('val_acc : ', model.evaluate([testS, testQ], testA, verbose=0)[1])
    print('===============================')
```
